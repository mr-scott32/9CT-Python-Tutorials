### Extension: Using Lists for More Complicated Purposes

In the following program below:

* rows represents the number of rows in the grid.
* cols represents the number of columns in the grid.
* We use a list comprehension to create a grid filled with '-' symbols initially.
* The for loop then iterates over each row in the grid and prints it, joining the elements of the row with spaces to create a visual representation of the grid.
* You can adjust the rows and cols variables to create grids of different sizes or modify the symbols used to represent the grid elements.


```python
# Define the dimensions of the grid
rows = 3
cols = 4

# Create a grid using a list of lists
grid = [['-' for _ in range(cols)] for _ in range(rows)]

# Print the grid
for row in grid:
    print(' '.join(row))
```

### Why use a Grid?
Whilst it may seem a bit farfetched to use a grid for this purpose, let's consider the following program that allows one to move around the grid, and which marks the location with 'X'.

This kind of grid would allow us to simulate simple games such as text-based adventures or Connect 4.

Have a look below:

```python
# Define the dimensions of the grid
rows = 3
cols = 3

# Create a grid using a list of lists
grid = [['-' for _ in range(cols)] for _ in range(rows)]

# Initial position of 'X'
x_row = 2
x_col = 2
grid[x_row][x_col] = 'X'

# Print the grid
def print_grid():
    for row in grid:
        print(' '.join(row))

# Print the initial grid with 'X'
print("Initial Grid:")
print_grid()

# Movement function
def move_x(direction):
    global x_row, x_col
    grid[x_row][x_col] = '-'  # Clear current position

    # Update position based on direction
    if direction == 'left' and x_col > 0:
        x_col -= 1
    elif direction == 'right' and x_col < cols - 1:
        x_col += 1
    elif direction == 'up' and x_row > 0:
        x_row -= 1
    elif direction == 'down' and x_row < rows - 1:
        x_row += 1

    grid[x_row][x_col] = 'X'  # Set new position
    print("\nAfter Move:")
    print_grid()

# Allow user to choose direction and move 'X'
while True:
    direction = input("\nEnter direction (left/right/up/down) or 'exit' to quit: ").lower()
    if direction == 'exit':
        break
    elif direction in ['left', 'right', 'up', 'down']:
        move_x(direction)
    else:
        print("Invalid direction! Please enter left/right/up/down or 'exit'.")
```
